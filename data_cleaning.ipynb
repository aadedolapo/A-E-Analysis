# Data Cleaning

### Clean the `Age_Group` column

```python
csv_file['Age_Group']
```

```python
# replacing 'jun-18' to 05-18
csv_file['Age_Group'].replace('Jun-18', '05-18', inplace=True)
```

```python
csv_file['Age_Group'].unique()
```

```python
csv_file['Chief_Complaint'].unique()
```

```python
csv_file['Chief_Complaint'].nunique()
```

```python
def wrangler(file):
    df = pd.read_csv(file)

    # replacing 'jun-18' to 05-18
    df['Age_Group'].replace('Jun-18', '05-18', inplace=True)
    
    df.reset_index(inplace=True)

    df = df.set_axis(['id', 'Place', 'Provider_Name', 'Age_Group', 'GP_Practice_Name',
       'Financial_Year', 'Financial_Month', 'Time_In_A&E', 'Arrival_Mode',
       'Chief_Complaint', 'Discharge_Destination', 'Attendance_Count'], axis='columns')

    return df
```

```python
clean_data = wrangler('AAEDataset.csv')
```

```python
clean_data
```

```python
clean_data.to_csv('AAEDatasetCleaned.csv', index=False)
```

### Q2 For practice 5, what is the percentage increase or decrease in attendances between the 2 years?

```python
gp_5 = clean_data.query("GP_Practice_Name=='GP Practice 5'")
```

```python
gp_5.groupby(['Financial_Year'])['Attendance_Count'].sum()
```

```python
gp_5['Attendance_Count'].sum()
```

```python
gp_5.loc[gp_5['Financial_Year'] == '2022-23', 'Attendance_Count'].sum()
```

```python
gp_5[gp_5['Financial_Year'] == '2022-23']['Attendance_Count'].sum()
```

```python

# Calculate the percentage change
attendance_year_1 = gp_5.loc[gp_5['Financial_Year'] == '2022-23', 'Attendance_Count'].sum()
attendance_year_2 = gp_5.loc[gp_5['Financial_Year'] == '2023-24', 'Attendance_Count'].sum()

percentage_change = ((attendance_year_2 - attendance_year_1) / attendance_year_1) * 100

print(f"Percentage Change: {percentage_change:.2f}%")
```
